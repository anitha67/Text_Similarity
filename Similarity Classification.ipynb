```python
def regex_sub(text):
    #ip_pattern = r'((((([0-9]|[1-9][0-9]|1[0-9]{2}|2[0-4][0-9]|25[0-5])\.){3}([0-9]|[1-9][0-9]|1[0-9]{2}|2[0-4][0-9]|25[0-5]))([^0-9]|$))|(((([0-9A-Fa-f]{1,4}:){7}([0-9A-Fa-f]{1,4}|:))|(([0-9A-Fa-f]{1,4}:){6}(:[0-9A-Fa-f]{1,4}|((25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)(\.(25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)){3})|:))|(([0-9A-Fa-f]{1,4}:){5}(((:[0-9A-Fa-f]{1,4}){1,2})|:((25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)(\.(25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)){3})|:))|(([0-9A-Fa-f]{1,4}:){4}(((:[0-9A-Fa-f]{1,4}){1,3})|((:[0-9A-Fa-f]{1,4})?:((25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)(\.(25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)){3}))|:))|(([0-9A-Fa-f]{1,4}:){3}(((:[0-9A-Fa-f]{1,4}){1,4})|((:[0-9A-Fa-f]{1,4}){0,2}:((25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)(\.(25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)){3}))|:))|(([0-9A-Fa-f]{1,4}:){2}(((:[0-9A-Fa-f]{1,4}){1,5})|((:[0-9A-Fa-f]{1,4}){0,3}:((25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)(\.(25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)){3}))|:))|(([0-9A-Fa-f]{1,4}:){1}(((:[0-9A-Fa-f]{1,4}){1,6})|((:[0-9A-Fa-f]{1,4}){0,4}:((25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)(\.(25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)){3}))|:))|(:(((:[0-9A-Fa-f]{1,4}){1,7})|((:[0-9A-Fa-f]{1,4}){0,5}:((25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)(\.(25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)){3}))|:)))(%.+)?([^A-Za-z0-9]|$)))'
    yyyy_date_pattern = r'([1-2][0-9]{3}[-\.\/])((([1-9]|1[0-2]|0[1-9])[-\.\/]([1-9]|0[1-9]|1[0-9]|2[0-9]|3[0-1])[^0-9])|(([1-9]|0[1-9]|1[0-9]|2[0-9]|3[0-1])[-\.\/]([1-9]|0[1-9]|1[0-2])[^0-9]))'
    m_d_y_data_pattern = r'[^0-9]((([1-9]|0[1-9]|1[0-2])[-\.\/]([1-9]|0[1-9]|1[0-9]|2[0-9]|3[0-1]))|(([1-9]|0[1-9]|1[0-9]|2[0-9]|3[0-1])[-\.\/]([1-9]|1[0-2]|0[1-9])))[-\.\/]([1-2][0-9]{3}|[0-9]{2})'
    timestamp_pattern_1 = r'([1-2][0-9]{3}[-\.\/])((([1-9]|1[0-2]|0[1-9])[-\.\/]([1-9]|0[1-9]|1[0-9]|2[0-9]|3[0-1])[^0-9])|(([1-9]|0[1-9]|1[0-9]|2[0-9]|3[0-1])[-\.\/]([1-9]|0[1-9]|1[0-2])))[\s]([0-9]|[0-1][0-9]|2[0-3]):([0-9]|[0-5][0-9]):([0-9]|[0-5][0-9])[^0-9]'
    timestamp_pattern_2 = r'[^0-9]((([1-9]|0[1-9]|1[0-2])[-\.\/]([1-9]|0[1-9]|1[0-9]|2[0-9]|3[0-1]))|(([1-9]|0[1-9]|1[0-9]|2[0-9]|3[0-1])[-\.\/]([1-9]|1[0-2]|0[1-9])))[-\.\/]([1-2][0-9]{3}|[0-9]{2})[\s]([0-9]|[0-1][0-9]|2[0-3]):([0-9]|[0-5][0-9]):([0-9]|[0-5][0-9])[^0-9]'
    #windows_file_path = r'[A-Za-z]:[\\][^\s]*'
    unix_file_path = r''
    #url_pattern = r'http[s]?:[\/]{2}[^\s]+'
    #email_pattern = r'[\S]+@[^\s\.]+\.[\S]+'
    html_tag = r'\<[^\>]*\>'
    text = re.sub(html_tag,' ',str(text))
    #text = re.sub(windows_file_path,' filepath ',text)
    #text = re.sub(url_pattern,' url ',text)
    #text = re.sub(email_pattern,' email ',text)
    #text = re.sub(ip_pattern,' ipaddress ',text)
    text = re.sub(timestamp_pattern_1,' timestamp ',text)
    text = re.sub(timestamp_pattern_2,' timestamp ',text)
    text = re.sub(yyyy_date_pattern,' timestamp ',text)
    text = re.sub(m_d_y_data_pattern,' timestamp ',text)
    text = text.lower()
    text = re.sub('[^a-z0-9\s]',' ',text)
    return text
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# Load the dataset
df = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/10kcases.csv")
df['description'] = df['description'].apply(lambda x: regex_sub(x))

# Split the dataset into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Convert the text in the column to a matrix of features
vectorizer = TfidfVectorizer()
train_features = vectorizer.fit_transform(train_df['description'])

# Train a text similarity model
model = NearestNeighbors(n_neighbors=10)
model.fit(train_features)

# Convert the input text from the query to a feature matrix
query_text = '''
<b> customer </b> 26464214-b614-4738-5e2f-704d5daaff2e <br><b> tenant_id </b> 26464214-b614-4738-5e2f-704d5daaff2e <br><b> tenant_name </b> travsport <br><b> agent_uuid </b> 83bca328-fb5e-4d88-4bb3-2f68dc45f305 <br><b> agent_hostname </b> TR-G90491QW <br><b> uuid </b> f0c4205f-ed89-435b-96aa-db89ef722e86 <br><b> event_type </b> ALARM <br><b> start_timestamp </b> 2020-09-23T08:55:14.255Z <br><b> alarm_type </b> AGENT_POLICY <br><b> juid </b>  <br><b> username </b>  <br><b> score </b> 50 <br><b> total_score </b> 100 <br><b> SENSORS </b> <br><b> customer </b> 26464214-b614-4738-5e2f-704d5daaff2e <br><b> tenant_id </b> 26464214-b614-4738-5e2f-704d5daaff2e <br><b> tenant_name </b>  <br><b> agent_uuid </b> 83bca328-fb5e-4d88-4bb3-2f68dc45f305 <br><b> agent_hostname </b>  <br><b> uuid </b> 4cc0d352-6a84-42d5-4be9-2e27427884d3 <br><b> event_type </b>  <br><b> timestamp </b> 2020-09-23T08:55:14.255765873Z <br><b> sensor_type </b> AGENT_POLICY <br><b> juid </b>  <br><b> username </b>  <br><b> score </b> 50 <br><b> created_by </b> policy:///2f7cdfb9-708e-4fb0-505b-be9d5e934464/5d697f10-76df-4e08-6308-86f1b50a0748?name=Potential+privilege+escalation+attack+detected <br><b> tags </b> ['AttackIndicators', 'KnownAttacks', 'PrivilegeEscalation', 'privilegeescalation', 'attackindicator'] <br><b> anonymised_description </b> [REDACTED] run by [REDACTED] with admin account privileges <br><b> customer </b> 26464214-b614-4738-5e2f-704d5daaff2e <br><b> tenant_id </b> 26464214-b614-4738-5e2f-704d5daaff2e <br><b> tenant_name </b>  <br><b> agent_uuid </b> 83bca328-fb5e-4d88-4bb3-2f68dc45f305 <br><b> agent_hostname </b>  <br><b> uuid </b> 0f42b017-ab86-4bff-475f-bcd1fcc0aaf9 <br><b> event_type </b>  <br><b> timestamp </b> 2020-09-23T09:00:19.112760478Z <br><b> sensor_type </b> AGENT_POLICY <br><b> juid </b>  <br><b> username </b>  <br><b> score </b> 50 <br><b> created_by </b> policy:///2f7cdfb9-708e-4fb0-505b-be9d5e934464/5d697f10-76df-4e08-6308-86f1b50a0748?name=Potential+privilege+escalation+attack+detected <br><b> tags </b> ['AttackIndicators', 'KnownAttacks', 'PrivilegeEscalation', 'privilegeescalation', 'attackindicator'] <br><b> anonymised_description </b> [REDACTED] run by [REDACTED] with admin account privileges <br>
'''
query_text = regex_sub(query_text)
query_features = vectorizer.transform([query_text])

# Use the trained model to find the k-nearest neighbors and their distances
distances, indices = model.kneighbors(query_features)

# Create a DataFrame of the nearest neighbors with their IDs and similarity scores
nearest_neighbors = train_df.iloc[indices[0]]
nearest_neighbors['similarity_score'] = 1 - distances.flatten()

# Select only the columns for ID, text, and similarity score
result_df = nearest_neighbors[['id', 'description', 'similarity_score']]

print(result_df)
```
